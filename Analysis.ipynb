                                            Sampled data
                                           

```python
spl = pd.DataFrame({'Man': [289807,141781,176298],
                   'Woman': [322191, 147048, 226985]},
                  index=(['18-25', '26-44','45-62']))
```

```python
spl
```

```python
spl.dtypes
```

```python
m = spl['Man'].sum()
w = spl['Woman'].sum()
print(f'Man: {m}\nWoman: {w}\nSum: {m+w}')
```

```python
spl.plot(kind = 'bar',figsize=(6,8),color = ['blue','pink'])
plt.title('Population distribution by age and gender in city Almaty')
plt.xlabel('Age Group')
plt.ylabel('Population')
plt.xticks(rotation=0)
plt.legend(title='Gender')
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.tight_layout()
plt.savefig('population_chart.png',dpi=300)
plt.show()
```
