```python
filtered_data = df[(df['2. Жынысыңыз\n2. Ваш пол?'] == 'Әйел адам / Женщина') &
                            (df['1. Сіздің жасыңыз?\n1. Ваш возраст?\n'] == '18-35')]

# Analyze responses to question 16
question_16_responses = filtered_data['16. Жақын туыстарыңыздың (балаңыз, ата-анаңыз, бауырыңыз, жұбайыңыз т.б.)  өмірін құтқару үшін,  ағза доноры болуға қаншалықты дайынсыз?\nНасколько вы готовы стать донором органа, если бы это было нужно для спасения жизни (ваших детей, родителей, супруга/с'].value_counts()

# Plot the responses
plt.figure(figsize=(10, 6))
sns.barplot(x=question_16_responses.index, y=question_16_responses.values)
plt.title('18-35 жас аралығындағы әйел адамдардың ағза доноры болуға дайындығы')
plt.xlabel('Жауаптар')
plt.ylabel('Саны')
plt.xticks(rotation=45, ha='center')  # Align text perfectly under bars
plt.show()
```

```python
df.rename(columns={'16. Жақын туыстарыңыздың (балаңыз, ата-анаңыз, бауырыңыз, жұбайыңыз т.б.)  өмірін құтқару үшін,  ағза доноры болуға қаншалықты дайынсыз?\nНасколько вы готовы стать донором органа, если бы это было нужно для спасения жизни (ваших детей, родителей, супруга/с':'16.Насколько вы готовы стать донором органа'},inplace = True)
```

```python
columns_of_interest = [
    '1. Сіздің жасыңыз?\n1. Ваш возраст?\n',
    '2. Жынысыңыз\n2. Ваш пол?',
    '16.Насколько вы готовы стать донором органа'
]

sub_df = df[columns_of_interest]
```

```python
df['1. Сіздің жасыңыз?\n1. Ваш возраст?\n'].unique()
```

```python
df['2. Жынысыңыз\n2. Ваш пол?'].unique()
```

```python
df['16.Насколько вы готовы стать донором органа'].unique()
```

```python

# Map age groups to approximate numeric values
age_mapping = {
    '18-35': 26.5,
    '36-44': 40,
    '45-62': 52,
    '60-74': 67,
    '75-90': 82.5
}
sub_df['1. Сіздің жасыңыз?\n1. Ваш возраст?\n'] = sub_df['1. Сіздің жасыңыз?\n1. Ваш возраст?\n'].map(age_mapping)
```

```python
# Map gender to numeric values
gender_mapping = {
    'Әйел адам / Женщина': 0,
    'Ер адам / Мужчина': 1
}
sub_df['2. Жынысыңыз\n2. Ваш пол?'] = sub_df['2. Жынысыңыз\n2. Ваш пол?'].map(gender_mapping)
```

```python
# Map string responses to numeric values for question 16
response_mapping_16 = {
    'Әрқашан дайынмын / Я всегда готов/а': 5,
    'Дайын сияқтымын / Я чувствую себя готовым': 4,
    'Моральды тұрғыда дайын емеспін / Я морально не готов/а': 3,
    'Мүлдем дайын емеспін / Я совсем не готов/а': 2,
    'Жауап беруге қиналамын / Я затрудняюсь ответить': 1
}
sub_df['16.Насколько вы готовы стать донором органа'] = sub_df['16.Насколько вы готовы стать донором органа'].map(response_mapping_16)
```

```python
sub_df
```

```python
# Drop rows with NaN values to ensure accurate correlation analysis
sub_df = sub_df.dropna()

# Correlation analysis
correlation_matrix = sub_df.corr(method='pearson')


# Display the correlation matrix
print("Pearson Correlation Matrix:")
print(correlation_matrix)

# Optionally, plot a heatmap for better visualization
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
import folium

# Центр карты (примерно центр Евразии для удобного отображения стран-партнеров)
map_center = [30, 70]  # Координаты широты и долготы

# Список стран и их координат
selected_countries_coords = {
    "United States": [37.0902, -95.7129],
    "Turkey": [38.9637, 35.2433],
    "Azerbaijan": [40.1431, 47.5769],
    "Lithuania": [55.1694, 23.8813],
    "Latvia": [56.8796, 24.6032],
    "Estonia": [58.5953, 25.0136],
    "India": [20.5937, 78.9629],
    "Kyrgyzstan": [41.2044, 74.7661],
    "Russia": [61.5240, 105.3188],
    "South Korea": [35.9078, 127.7669],
    "Kazakhstan": [48.0196, 66.9237]
}

# Создание карты
m = folium.Map(location=map_center, zoom_start=2)

# Добавление маркеров для стран
for country, coords in selected_countries_coords.items():
    folium.Marker(location=coords, popup=country, icon=folium.Icon(color="blue")).add_to(m)

# Сохранение карты в файл
map_file_path = "selected_map_partners.html"
m.save(map_file_path)

print(f"Карта сохранена в файл: {map_file_path}")
```
